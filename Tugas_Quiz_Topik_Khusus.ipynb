# **TUGAS 1**
**PROGRAM KALKULATOR**
*   Menu: Program menampilkan menu pilihan operator (1-8) di awal dan setelah setiap perhitungan (menggunakan while loop untuk pengulangan opsional, mirip for looping).
*   Input Variabel: Menggunakan float(input()) untuk mendukung input string yang dikonversi ke int/float (fleksibel untuk desimal).
*   Operator: Dipilih melalui input string pilihan, lalu diterjemahkan ke operator matematika (+, -, *, /, //, %, **).
*   If Conditional: Digunakan untuk memilih operasi berdasarkan pilihan, serta menangani error seperti pembagian nol.
*   Hitung Operasi & Tampilkan Hasil: Perhitungan dilakukan berdasarkan operator, hasil ditampilkan dengan format sederhana.
*   Error Handling: Menangani input tidak valid dan pembagian/modulo dengan nol menggunakan try-except dan if conditional.
*   Optional Looping: While loop memungkinkan pengulangan tanpa for eksplisit, tapi bisa dihentikan kapan saja.

```python
# Program Kalkulator Sederhana dengan Python
# Mendukung operasi: +, -, *, /, //, %, **
# Menggunakan input (float untuk fleksibilitas), operator (string), if conditional
# Optional: Menggunakan while loop untuk menu utama agar bisa diulang (mirip for looping)

def kalkulator():
    print("=== SELAMAT DATANG DI KALKULATOR ===")
    print("Pilih operator yang diinginkan:")
    print("1. Penjumlahan (+)")
    print("2. Pengurangan (-)")
    print("3. Perkalian (*)")
    print("4. Pembagian (/)")
    print("5. Pembagian Bulat (//)")
    print("6. Modulo (%)")
    print("7. Pangkat (**)")
    print("8. Keluar")

    while True:
        pilihan = input("\nMasukkan pilihan (1-8): ").strip()

        if pilihan == '8':
            print("Terima kasih telah menggunakan kalkulator!")
            break

        if pilihan in ['1', '2', '3', '4', '5', '6', '7']:
            try:
                # Input variabel (menggunakan float untuk mendukung int/float)
                num1 = float(input("Masukkan angka pertama: "))
                num2 = float(input("Masukkan angka kedua: "))

                # Tentukan operator berdasarkan pilihan
                if pilihan == '1':
                    operator = '+'
                    hasil = num1 + num2
                elif pilihan == '2':
                    operator = '-'
                    hasil = num1 - num2
                elif pilihan == '3':
                    operator = '*'
                    hasil = num1 * num2
                elif pilihan == '4':
                    operator = '/'
                    if num2 == 0:
                        print("Error: Pembagian dengan nol tidak diperbolehkan!")
                        continue
                    hasil = num1 / num2
                elif pilihan == '5':
                    operator = '//'
                    if num2 == 0:
                        print("Error: Pembagian dengan nol tidak diperbolehkan!")
                        continue
                    hasil = num1 // num2
                elif pilihan == '6':
                    operator = '%'
                    if num2 == 0:
                        print("Error: Modulo dengan nol tidak diperbolehkan!")
                        continue
                    hasil = num1 % num2
                elif pilihan == '7':
                    operator = '**'
                    hasil = num1 ** num2

                # Tampilkan hasil
                print(f"\nHasil: {num1} {operator} {num2} = {hasil}")

            except ValueError:
                print("Error: Masukkan angka yang valid!")
            except Exception as e:
                print(f"Terjadi kesalahan: {e}")

        else:
            print("Pilihan tidak valid! Silakan pilih 1-8.")

        # Opsional: Loop ulang untuk menu (menggantikan for looping sederhana)
        ulang = input("\nHitung lagi? (yes/no): ").strip().lower()
        if ulang != 'y':
            print("Terima kasih!")
            break

# Jalankan program
if __name__ == "__main__":
    kalkulator()
```

# **TUGAS 2**

**PROGRAM PERHITUNGAN BANGUN DATAR ATAU BANGUN RUANG**
*   Menu: Dibuat menggunakan print dan input untuk pilihan utama (1-5) dan sub-pilihan untuk setiap bangun.
*   Input Variables: Menggunakan input() dengan tipe string untuk menu, dan float(input()) untuk variabel ukuran (mendukung int/float). Contoh: alas, tinggi, jari-jari, dll.
*   Perhitungan:
    1.   Luas/Keliling segitiga dan lingkaran menggunakan rumus standar dengan operator aritmatika (*, +, 0.5 untuk pembagian).
    2.   Volume/Luas permukaan kubus dan balok juga menggunakan operator (* untuk perkalian).
    3.   Menggunakan math.pi untuk lingkaran.
*   Tampilan Hasil: Hasil di-print dengan format sederhana menggunakan f-string.
*   If Conditional: Digunakan untuk memeriksa pilihan menu, sub-pilihan, dan validasi input.
*   For (Looping): Opsional, menggunakan while True untuk loop utama agar program bisa diulang hingga user pilih keluar (5).




```python
import math  # Untuk menggunakan nilai pi pada lingkaran

# Fungsi untuk menghitung luas segitiga
def luas_segitiga(alas, tinggi):
    return 0.5 * alas * tinggi

# Fungsi untuk menghitung keliling segitiga (asumsi segitiga siku-siku, input ketiga sisi)
def keliling_segitiga(sisi1, sisi2, sisi3):
    return sisi1 + sisi2 + sisi3

# Fungsi untuk menghitung luas lingkaran
def luas_lingkaran(r):
    return math.pi * r * r

# Fungsi untuk menghitung keliling lingkaran
def keliling_lingkaran(r):
    return 2 * math.pi * r

# Fungsi untuk menghitung volume kubus
def volume_kubus(sisi):
    return sisi * sisi * sisi

# Fungsi untuk menghitung luas permukaan kubus
def luas_permukaan_kubus(sisi):
    return 6 * sisi * sisi

# Fungsi untuk menghitung volume balok
def volume_balok(p, l, t):
    return p * l * t

# Fungsi untuk menghitung luas permukaan balok
def luas_permukaan_balok(p, l, t):
    return 2 * (p * l + p * t + l * t)

# Program utama
def main():
    while True:  # For loop opsional untuk mengulang program hingga user keluar
        print("\n=== PROGRAM PERHITUNGAN BANGUN DATAR ATAU BANGUN RUANG ===")
        print("Pilih Bangun Datar/Ruang yang Ingin Dihitung: ")
        print("1. Segitiga (Luas/Keliling)")
        print("2. Lingkaran (Luas/Keliling)")
        print("3. Kubus (Volume/Luas Permukaan)")
        print("4. Balok (Volume/Luas Permukaan)")
        print("5. Keluar")

        pilihan = input("Pilih menu (1-5): ")  # Input string untuk pilihan menu

        if pilihan == '5':  # If conditional untuk keluar
            print("Terima kasih telah menggunakan program ini!")
            break

        elif pilihan in ['1', '2', '3', '4']:  # If conditional untuk pilihan valid
            if pilihan == '1':  # Segitiga
                print("\n--- SEGITIGA ---")
                sub_pilihan = input("Pilih: 1. Luas | 2. Keliling: ")
                if sub_pilihan == '1':  # Luas segitiga
                    alas = float(input("Masukkan alas: "))  # Input float
                    tinggi = float(input("Masukkan tinggi: "))
                    hasil = luas_segitiga(alas, tinggi)
                    print(f"Luas segitiga = {hasil}")
                elif sub_pilihan == '2':  # Keliling segitiga
                    sisi1 = float(input("Masukkan sisi 1: "))
                    sisi2 = float(input("Masukkan sisi 2: "))
                    sisi3 = float(input("Masukkan sisi 3: "))
                    hasil = keliling_segitiga(sisi1, sisi2, sisi3)
                    print(f"Keliling segitiga = {hasil}")
                else:
                    print("Pilihan tidak valid!")

            elif pilihan == '2':  # Lingkaran
                print("\n--- LINGKARAN ---")
                sub_pilihan = input("Pilih: 1. Luas | 2. Keliling: ")
                r = float(input("Masukkan jari-jari: "))  # Input float
                if sub_pilihan == '1':
                    hasil = luas_lingkaran(r)
                    print(f"Luas lingkaran = {hasil}")
                elif sub_pilihan == '2':
                    hasil = keliling_lingkaran(r)
                    print(f"Keliling lingkaran = {hasil}")
                else:
                    print("Pilihan tidak valid!")

            elif pilihan == '3':  # Kubus
                print("\n--- KUBUS ---")
                sub_pilihan = input("Pilih: 1. Volume | 2. Luas Permukaan: ")
                sisi = float(input("Masukkan panjang sisi: "))  # Input float
                if sub_pilihan == '1':
                    hasil = volume_kubus(sisi)
                    print(f"Volume kubus = {hasil}")
                elif sub_pilihan == '2':
                    hasil = luas_permukaan_kubus(sisi)
                    print(f"Luas permukaan kubus = {hasil}")
                else:
                    print("Pilihan tidak valid!")

            elif pilihan == '4':  # Balok
                print("\n--- BALOK ---")
                sub_pilihan = input("Pilih: 1. Volume | 2. Luas Permukaan: ")
                p = float(input("Masukkan panjang: "))  # Input float
                l = float(input("Masukkan lebar: "))
                t = float(input("Masukkan tinggi: "))
                if sub_pilihan == '1':
                    hasil = volume_balok(p, l, t)
                    print(f"Volume balok = {hasil}")
                elif sub_pilihan == '2':
                    hasil = luas_permukaan_balok(p, l, t)
                    print(f"Luas permukaan balok = {hasil}")
                else:
                    print("Pilihan tidak valid!")

            # Operator digunakan dalam fungsi-fungsi perhitungan (misal *, +, 0.5 untuk pembagian)

        else:
            print("Pilihan tidak valid! Silakan pilih 1-5.")

# Jalankan program
if __name__ == "__main__":
    main()
```

# **TUGAS 3**
**PROGRAM KONVERSI SUHU**
*   Menu: Dibuat menggunakan print dan input untuk pilihan utama (1-5).
*   Menggunakan input() dengan tipe string untuk menu, dan float(input()) untuk nilai suhu (mendukung int/float). Contoh: celsius, fahrenheit, reamur, dll.
*   Perhitungan: Setiap konversi (C ke F, C ke R, F ke C, R ke C) menggunakan fungsi terpisah dengan rumus standar dan operator aritmatika (* untuk perkalian, / untuk pembagian, + dan - sesuai rumus).
*   Tampilan Hasil: Hasil di-print dengan format sederhana menggunakan f-string, menampilkan input dan output suhu.
*   If Conditional: Digunakan untuk memeriksa pilihan menu dan validasi input (jika tidak 1-5, tampilkan pesan error).
*   For (Looping): Opsional, menggunakan while True untuk loop utama agar program bisa diulang hingga user pilih keluar (5).

```python
# Program Konversi Suhu
# Rumus:
# C to F: F = (C * 9/5) + 32
# C to R: R = C * 4/5
# F to C: C = (F - 32) * 5/9
# R to C: C = R * 5/4

# Fungsi konversi Celsius ke Fahrenheit
def c_to_f(celsius):
    return (celsius * 9/5) + 32

# Fungsi konversi Celsius ke Reamur
def c_to_r(celsius):
    return celsius * 4/5

# Fungsi konversi Fahrenheit ke Celsius
def f_to_c(fahrenheit):
    return (fahrenheit - 32) * 5/9

# Fungsi konversi Reamur ke Celsius
def r_to_c(reamur):
    return reamur * 5/4

# Program utama
def main():
    while True:  # For loop opsional menggunakan while untuk mengulang program hingga user keluar
        print("\n=== PROGRAM KONVERSI SUHU ===")
        print("Pilih Konversi: ")
        print("1. Celsius ke Fahrenheit")
        print("2. Celsius ke Reamur")
        print("3. Fahrenheit ke Celsius")
        print("4. Reamur ke Celsius")
        print("5. Keluar")

        pilihan = input("Pilih menu (1-5): ")  # Input string untuk pilihan menu

        if pilihan == '5':  # If conditional untuk keluar
            print("Terima kasih telah menggunakan program ini!")
            break

        elif pilihan in ['1', '2', '3', '4']:  # If conditional untuk pilihan valid
            if pilihan == '1':  # Celsius ke Fahrenheit
                print("\n--- CELSIUS KE FAHRENHEIT ---")
                celsius = float(input("Masukkan derajat Celsius: "))  # Input float
                hasil = c_to_f(celsius)
                print(f"Hasil konversi: {celsius}°C = {hasil}°F")
                # Operator digunakan: *, /, + dalam fungsi

            elif pilihan == '2':  # Celsius ke Reamur
                print("\n--- CELSIUS KE REAMUR ---")
                celsius = float(input("Masukkan derajat Celsius: "))  # Input float
                hasil = c_to_r(celsius)
                print(f"Hasil konversi: {celsius}°C = {hasil}°R")
                # Operator digunakan: *, / dalam fungsi

            elif pilihan == '3':  # Fahrenheit ke Celsius
                print("\n--- FAHRENHEIT KE CELSIUS ---")
                fahrenheit = float(input("Masukkan derajat Fahrenheit: "))  # Input float
                hasil = f_to_c(fahrenheit)
                print(f"Hasil konversi: {fahrenheit}°F = {hasil}°C")
                # Operator digunakan: -, *, / dalam fungsi

            elif pilihan == '4':  # Reamur ke Celsius
                print("\n--- REAMUR KE CELSIUS ---")
                reamur = float(input("Masukkan derajat Reamur: "))  # Input float
                hasil = r_to_c(reamur)
                print(f"Hasil konversi: {reamur}°R = {hasil}°C")
                # Operator digunakan: *, / dalam fungsi

        else:
            print("Pilihan tidak valid! Silakan pilih 1-5.")

# Jalankan program
if __name__ == "__main__":
    main()
```
